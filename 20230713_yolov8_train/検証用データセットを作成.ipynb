```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import random, shutil
from pathlib import Path
from ultralytics.yolo.engine.results import Results

import utils

config_path = Path("./dataset/mask_yolo_dataset/config.yaml")
yolo_data_train = utils.YOLO_dataset(str(config_path), "train")

validation_dataset_dir = config_path.parent / "valid"
if validation_dataset_dir.exists():
    raise Exception("validation_dataset_dir already exists.")
validation_dataset_dir.mkdir(exist_ok=True)
(validation_dataset_dir/"images").mkdir()
(validation_dataset_dir/"labels").mkdir()

for i in random.sample(range(len(yolo_data_train)), 100):
    yolo_data = yolo_data_train[i]

    img_path = Path(yolo_data["img_path"])
    label_path = Path(yolo_data["label_path"])

    shutil.move(img_path, validation_dataset_dir/"images")
    shutil.move(label_path, validation_dataset_dir/"labels")
```
