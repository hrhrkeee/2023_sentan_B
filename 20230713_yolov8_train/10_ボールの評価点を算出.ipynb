```python
import math

def darts_score_calculate(  darts_distance:float,                              # ボードの中心とダーツの距離[px]
                            darts_sita:float,                                  # ボードの中心座標からダーツまでの角度[rad]
                            board_radius:int  ,                                # ボードの半径[px]
                            each_circle_radius:list[int] = [0, 20, 60, 100],   # ボードの各円の半径[%]
                            radius_divisions:int = 3,                          # ボードの円の数
                            circle_divisions:int = 12,                         # ボードの円の分割数
                            score_table:list=None,                             # ボードの点数表
                        ):
    
    score = 0
    base_angle = float(360/circle_divisions) 

    n_darts_distance = float(darts_distance/board_radius)
    n_each_circle_radius = [float(i/100) for i in each_circle_radius]

    # score_tableの中の配列の長さをcircle_divisionsまで拡張する
    score_table = [(table*(math.ceil(circle_divisions/len(table))))[:12] for table in score_table]
        
    for j in range(radius_divisions) :
        if (n_each_circle_radius[j] <= n_darts_distance < n_each_circle_radius[j+1]) : # ダーツの位置を、rから判断
            print("darts_distanceは{}番目の円の範囲にありました。" .format(j+1))  
        
            for k in range(circle_divisions) : 
                if(base_angle * k <= darts_sita < base_angle * (k+1)) :  # ダーツの位置を、θから判断
                    print("θは{}°以内の位置にありました。" .format(base_angle * (k+1)))             
                    return score_table[j][k]          

    return score






score_table1 = [200]
score_table2 = [110, 150, 100, 170, 90, 190, 80, 180, 130, 160, 120, 140]
score_table3 = [40, 10, 60, 10, 20, 10, 70, 10, 30, 10, 50, 10]
score_table  = [score_table1, score_table2, score_table3]


darts_list = [
    {
        "ball_distance":10,  # [px]
        "ball_sita":20,     # [rad]
        "board_radius":100, # [px]
    },
    {
        "ball_distance":40,  # [px]
        "ball_sita":20,    # [rad]
        "board_radius":100, # [px]
    },
    {
        "ball_distance":90,  # [px]
        "ball_sita":20,    # [rad]
        "board_radius":100, # [px]
    },
]


for i, darts in enumerate(darts_list):
    print("{}本目の計算を行います。".format(i+1))
    
    score = darts_score_calculate(  darts_distance  = darts["ball_distance"],
                                    darts_sita      = darts["ball_sita"],
                                    board_radius    = darts["board_radius"],
                                    score_table     = score_table
                                )
    print(score)
    print()
    
```
